```python
import pandas as pd
import re
pd.set_option('display.max_colwidth',200)
```

```python
df = pd.read_csv(r'C:\Users\amhan\AnacondaProjects\chat_bot\tranning_data.csv', encoding='latin-1')
df
```

```python
convo = df.iloc[:,0]
convo
```

```python
clist = []
def qa_pairs(x):
    cpairs = re.findall(r": (.*?)(?:$|\\n)", x)
    clist.extend(list(zip(cpairs, cpairs[1:])))
convo.map(qa_pairs);
convo_frame = pd.Series(dict(clist)).to_frame().reset_index()
convo_frame.columns = ['q', 'a']
```
